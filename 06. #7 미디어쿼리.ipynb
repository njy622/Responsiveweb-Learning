# 7. 미디어쿼리

## 1) 미디어 쿼리
- 미디어 쿼리(media query)는 미디어 타입을 인식하고, 콘텐츠를 읽어들이는 기기나 브라우저의 물리적 속성을 감지할 수 있는 유용한 장치(기능) 입니다.
- 모든 미디어 쿼리는 다음 두가지 구성 요소를 지니고 있습니다.
    - 미디어 타입
    - 조건에 대한 물음(쿼리)




```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>미디어쿼리</title>
</head>
<style>
    img{
        width: 200px;
        height: 200px;
    }
    @media screen and (min-width: 800px){
        img{
            width: 400px;
            height: 400px;
        }
    }
    @media screen and (min-width: 1200px){
        img{
            width: 800px;
            height: 800px;
        }
    }

    /*
        미디어 조건
        1) 미디어 타입 : 어떤 미디어에서 미디어를 사용할 것인가?
            - screen (많이 사용) : 브라우저 화면 
            - print : 인쇄결과물에 대한 쿼리를 지정하고 싶을때
            - speech : 웹문서를 음성을 통해 열람을 하는 상황
        2) 쿼리
            - 주로 사용하는 조건 : max-width(이하), min-width(이상)

        유의사항
        - 다중 미디어 조건을 줄때는, 조건을 줄때, 기존의 범위를 겹치지 않도록 정하기
        - width 조건을 줄 수 있으나 정확한 값에 일치해야만 적용됨으로 사용안함
    */
</style>
<body>
    <img src="data:image/jpeg;base64,/9j/4AAQSkZJRgABAQAAAQABAAD/2wCEAAkGBxMTEhUSExIVFRUXGBcXGBUXFxcXGhcXFxgYFxcXGhcdHSggGBolHRUVIjEhJSkrLi4uFx8zODMtNygtLisBCgoKDg0OGhAQFy0fHx0tLSstLS0tLS0tLS0tLS0tKy0tLS0tLS0tLS0tLS0tLS0tLS0tLS0tLS0xLS0tLSsuLf/AABEIANAA8wMBIgACEQEDEQH/xAAcAAAABwEBAAAAAAAAAAAAAAABAgMEBQYHAAj/xABEEAABAwICBwQHBAgFBQEAAAABAAIDBBEFIQYSMUFRYXEHEyKBMkKRobHB8CNSYoIUM2NykrLR4UNTosLxJDRzk9IX/8QAGgEAAwEBAQEAAAAAAAAAAAAAAQIDAAQFBv/EAC0RAAICAQQBAwIEBwAAAAAAAAABAhEDBBIhMUEFMlETYSKx0eEUFSMzQnGh/9oADAMBAAIRAxEAPwAoKOG3Sbfr64J9TRrpnKkTjGwIYLqTpYbIYmWS7QuVybLVRKULhwUvE1pGxV6B9lL00yAKHE1I07lHVGFg7lJmREEiwCuVGD8FE1NK5u0ZK/AAplW4eDuTqQrRTYBmpemCSqKEsOQy+velYHBPQGx4wIJJbb03lqQFF1lcnQlWPamtsoiprCd6ay1F0VoJSSypFY4znOJQxw5/8JZjE4hjXK5OTLpJIUpo0vO6wSsMSZVz87IxiK2Hpnb06102p25JYFOwBtZGiFygYxO6aNAI6pYU+iYobE9IKalH28rWnaGDxPPRgz89irFT2rwgkRUz322Oe5rL+QDkyi30I2aMI0q2NZnTdqpJANK23KQ38rtsr1o9pLBVjwEtfbON1gfI7HeSLxySujbkSequCWcEi9TQQpcuQaqBGwGYwtzUvSsyCY00WaloW5KuSVs0VwKtSgSQKVupjBmmyfUj1HEp1TuWATLXIr3IkRyVM0k7R6WC7Iv+okH3CBGDzksb/lB8kVFvoUuQmTyGe4WB4n2h1svovbC3hG0X83Oub9LKvVOKzyenNK/957nfEqix/LFs9K1rWfeb7QFWcROpmMxyzWCXJ3e5KRyOGwkdMlRQXyA2CauumpcXLOKfGZ27JX9Cbj2G4U5Q6XPb+sjDubfCfgQfYEkoSfTHjKKLeyJLNamOHYzDNYMfZx9V3hPluPkpRrFzOEk+S25UAxl0/pYk3hYpKmCtHGRlMVc2zVBTvu76Kl8QnsLKFj23+vJNJUCHI7Yfr+6XYmzT9fW5OoApFB1AxUbS3TsgmGjI4PmyOeeUf/17OKT7QtJi29HCbZfauBzz/wAMcMtvW3FZ4qwj5YjYq99yXOJc45kkkkniScyia6TcUAVbFFmykKZwrF3sIIOYNxtuDtuLb1AXSsD7FZSox6E0I0l/TITrfrY7B34gdjvcb9FYwFi/ZTWlla1l8pGuaedhrD+VbSclDKkpcBR1lySLlylQxQ6ZieXRaeLJC8J27GDNS9kjE36+tyVcUpjikqvEY4I3Syu1WN2njwAG9x4Ik9Q1oLnEBrQSXHYAMyVkOlekDqqW+YibcRs4DZrHi4+7Z1pCFiSdD7S3TiervG0mKD/LBzeOMjh6X7uzrtVTJXErgwqwgIKMEVoRgETA2Rm3QtR2hFIxwSgauCUATIALB5qzYLpJJHYPPeM5+kOjuHI+5V6NifQU3mmq+zGmUFYyVusx1wfIg8CNxUjHJZZnh9Y+J4c3LcRnZw4HPmrnDibZGBzehG9p4dOe9LKNci1yK109zZFhH1xSMeZTuNq5pPkvFB2BFxbEW00Ekxt4RkOLjk0e23vSzAqN2m4jd0dODsHeP6nJo8hc/mSpWwt8FIkmL3Oc43c4kk8STclJOchIXFXJhUKArljAhHYiWRmLGLn2cf8Af0375/lct8lCw3sog1q+I/dD3exhH+5blMclLN2gxGxQIpK5RHK+2GwTN+1TssHhUHJkUxkKA2SUkiBzk2qJg0FztjQSTyGZPsWoJTu0PGLAUzTts6TpfwN92t/Cs/en+LVZmkfI7a4k9BuHkLDyTCy6qpURu2Tuh+i0tdL3bLNa0az5CLhrd2XrOJyAvx4LQ58PwChPczlskosHa4kmcDzawFsZ5WBSGi1UaPApaqMDvXd44HbZ2v3LCf3Rn7VkT3kkkkkkkknMknaSd5JU+2MajpZolRyUhr8PcCxubmNJLdW4aSA7NhbtI4XWdtYr5okTBglbJJk2d2pED6xLQwuHHf8A+s8FSYmJ4CsJ3a7VTpka4xFUMNQlGoCFwKxh5ApvC2g2uAeu/h8FA0z8wrJgOAVFUT3DXENNicg1uW9xIbfPYDdPdKzEmKMPHXYRcc9luaYMcYJR9zIHfluI42Tyup6ujIbOwhp9F+1pPDWG/wChvSFaC5oJ22/tz4rLn7mZZqePfusOFs/inbWqE0Tq9eIsO2M22+qc2/Aj8qn2hcU1ToqjmtWQ6WVXeVczvxlo6Ms35LZWNWEVT9aV7uLnH2klHH2CYm5JkpVySKsyYVCuK4IBBCUjCTATiBixjUexWkvPLJuZHq+b3D5NK1mcZKgdi9IWwTSWyc9rR+UXP860CUKGV/iGQxLfrNcjlq5SGOmiu1VXEWWcrbe7VBYnS3VEBMhC5Qml0+pSyfisz2nP3AqZkjLTmq3p9lTM5yD+R6eK5QW+DOJUgU4lCQcFdki86EaWQRU8lFWNcYJNbMAmweLOaQ3xWvmCMwSjPw/AYnd539TUDaIGg2NtznljLDq4KiApWNLtTDZY9KNJX1bmNDBFDHlFC21mN4m2Rd5WAyG8mHaQENJTF5yyG88FOUuGxja254nP3bEs80MfB2abQ5dRyuF8kOyYc0cP63VmjjA2ADyCUELTtaD1AK5/41fB6f8AI3X9z/hTpHX/AOEm4q4S4RE/1dU8WnV92z3KGxTAJGAvb9o0C5sPE0DaSOA4jzsr49RGf2ODU+mZsC3dr7EYw5XWpV2mjMJpKWmiiEsr4myuu4ta3XuS42FyXO18ssh0WSudYFTWll56ekq2XIbEKWU7dSSIkt1uGs14I6KmTmjz0azovpVT41BLTTR93IG3dHfW8N7CWN1hmCRuyJG0FZ1VsdDJJBJm6NxYbZX1TkehAuk+xiJ5xNpbfVbHKZDu1SLAHq7V9iX7QKprsQqXNzGuG9SxjWO97SEMTqVIz6FtEai1SW3ye1zbcx4h55O9qv0TFl2jklqmE/tGjycQ0/FaxE1Jn91jQ6BbGsCxSAxzSMPqve32OIXoNrVk3aZg5iqe+A8Ew1r7g8Czh8D5lJB8hl0U690oymJReifUz7roXJMRFF5oDBZSscd/opN0X0OCfaYje5TiFiW7lGhZmhQTeezWm1MPhy9LWf7XG3uAVjlaozQ5lqKmH7Jh9ov81MPbkuKfuY66GBK5IvablclCJQTXSzoQ5Q1FPdT9G26o+BSLqcNvuVI7SsPIpA4D0ZWk9C14+JatXdGFXtKsL7+nlhG1zTq/vDxN/wBQCEZchPOMoSBCe1MdiRYg/WSaPauomEAR40VCx+qQeBB9hv8AJAK75LVQQBrQE+a1NIZgQCNhGSdMcvJyJ3yfaabaopRFGhKtKI16UMgU0jsHVM25AUvFQvaA8XttBVdZNZTEWNv1NS+SvBryQzKb9pQtJqEQzGwAY/MAbBxaOVz7CEGjeNyUj3FrWyRPFpIX5te3dfg4Z2PMqT0zddjDv1vi0/0CqjXL0MUlOPJ8lr8Kw53GJpkHaFTwxPZRULaeR20+ANDvvG2b7cDZUCWQuJJNySSSTcknMkniTc+aagpRipGKXRx2TGBm88I/ax/ztWyRtWRaIRa1XCODw7+AF3+1bAwKWZ8oeIZoUdpFgzKqB0Lsjta77rxsPTceRUoAusojHnrEKN8MjopAWuaSCPraN6SjksVrHaNo2J4jOxv2sYzA9dgzI6tzPS44LISbLohKyUlRYqWQEX3py5gKiMMl3KXOzLK66o8oUSkiGaQYM04mOQ/t5pq12aDCeiNEjeipj+yZ/KFMKC0JN6Gm/wDG1ToXmz9zKIRfTgm65LLkoSj4W3NWui2KvYZGrDDkE7MOXOUViEydzSZKAxGoWSAZzpzo05z3VELbg+KRgGYO94A2g7+BudmzPpGreoVX9IdDYai72Wik+8B4XH8TePMe9UjkrhgaMeckyVK4/o9UUp+1YdQmzZB4mOOZFjuNgcjY5FQ2uq2mKPaPEHR5bW8P6KapMXYd9jwOX9lV9dDdTnijI68Gty4eE7Rd2VF0q2dUaORzfRJHQ2TqPFJW+tfqAVzvS/DPTx+sL/JFw71HZKqm3HJeDP4T/VJ1GLSuFtaw/CLf3SrTSKy9XxpWrJDSSvD3BgNw29+p3eXzUOCiNRgu2EdsaPBz5pZsjnLyKtKWYkWJ1SQue4MYC5ziAANpJ3J7JFz7NKK8sk25jdUHm/8AsD/EFpLAonRzChTQNi2u9J54vO3yGQHIKYYuabt2UQYIbLguukCBZYt2hYH+jVJLRaOTxsHD7zfI+4hbUVSe1imDqRkls2SgA8ntdf3tang6YGuDK6OSxU/BJdlvP3qtQnNTlE/w8/r681242SFZSkQjPOSIDmizHoHQI3w+m/c+ZVhCgtCo9Whph+yafbn81OhedP3MqgVy5clCVjDwpdpyUXR7FIB2ScUQrpbBVuoku5SeKz7VXaqujhY6WV4YwbSeO4AbS7kEQolIQoDSTTSnpbsB72Yf4bDk0/jdsb0zPLeqHpNp5NPeOC8MWzL9Y8b9Zw9Ecm+ZOwU+yosfyK5ErpDpDPVv1pneEG7Y25MZ0G8/iNz8FDo6LZPQLC3QgoCuQMH10cOSQXAo2ChwG3SgiTO6UDzxRsI6EKNqJmZncSn+H09/E8np89qMbbA+B1hWFy1D+7hZrO27gABtJJystS0T0UZSDXcQ+YjNw2MB2tZfPqTt5LN4xqm7CQW5gjIg7iLHIq3YTppJGA2oHeN+8LB4+T93A81p45NcBTRfwlGpvR1TJWh7HazTsPyI3HkUuuVqigdCEAQhAwNlSu1icNo2t3vlaB+Vr3H4D2q6rGe0rHRUVPdsN44btB3OefTdzGQb+U8U8FbFZU4hmpKCTJR8eSeQC9l1xEHt8l0TC5waBmSAOpyCTcVZezjDe/r4ha7WHvXdGWI/1avtWlLizG8UMAjjYwbGNa32AD5JwCiopcvPKC10CS7xAiYgKUpzNJYJvA1ROl+kEVHD3khu45MYNr3cBwA3nd7AnSsVkRpdj8dMzWfmTcNYNrz8hsuefkscxrGZal+vI791o9Fg4AfPaUGMYnLUymWV13HdsDRua0bgEwV4xoW7OQrkKYwCEoQEYMRoAk4IpalwEVwQoIlqoAEo5F1UKNYVHaCcgE4paFz3BrWlzjkGtBLj0AzKu+C9m9ZJYlrIR+0d4v4W3I87Kc8mOHvkkMot9IhMG0Xc4a8jmDg0vaD1OfNP6jR6Rty0X8+G64uFbG9kEhGdYwHgInEe3WTDEOyysiGtFJHKeDSWO8r5e9NHVYfkGyRVDUap1XjVPPYeh3JKpkBIISmJd/Ge5qo3XHqvGfUO3jnmFGOGpmCSw7CdoPA8Oqupp9C0W7RzHnwOBGbTbXZfJw48nc1pmH1zJmCSN12nLmDvaRucOCxCmn3qTocYmgcXwP1Tvac2u5Obv6jPmlyY96tdjJ0bOEcBZV/+ozgWNLFrbzrPt/D/AHUJimnlbMC3vBE03yibqm3DXJLvYQub6bG3IuvaHpk2FrqaB15nAte9p/VNORAI/wAQjK3q7dtlkoXIwaqxjQrdikYUhTNtmmlOy6dOKoBBnOWy9j+Cd1TuqXDxTZN5RtOXtNz0AWY6I4GayobHnqDxSO4NHzOwdV6Fo2hrWtaLNAAA4AZAKOWXFDIdlIyFK3SDyucYTL0CKSuWMQ+K4jHSwvnlNmMF+bj6rW8XE7F580kx2WsndPKduTWA+GNm5jfmd5uVZO1LSf8ASqjuYnfYQkgW2PkzDn8x6o5XPrKjFdUI0ibYCBCVycwCMAuASgasA6NqOUYZBJyOTGOckyf+FxKXgh3lL2YIyHeVZNFNEZat2sPBEDnIQfNrB6zvcN/BLaHaMmrku64hb6bhtP4Gnid53DnZbJSwtjaGsaGtaLBoyAA2ADcvE9T9VWD+lj93l/H7nZp9Nv8AxS6GWAaNwUzbRMANrF5sXu6ut7hYclZKVg5JnA3W5BTFJT2GY+vkvJ0kJ5p73zflnTlkoqhWJosjGIJQBCvejiVUzicmQuN6Pw1LCyWNrxz2jm07WnosV030IkoryMu+nJsSdrL7A7iODuPDJehE2raNsjS1zQ4EEEEXBB2gjeEIqWF7sfXleDNqXDPJ19Q8jsPy6p3FPndWHtA0SdRTFouYJCTG456p26hPEe8cwVTmO3HaF6WPKpRUkRcWmPKtgOaZaqUJKLqlM3YoUI8TUdsaVYxAIqzIc04w6iknkbFG3Wc45D5ngOadYDgE1U7Vib4R6Tzk1vU7zyGa1jRzR6KkZZniefSkO13Tg3kllkoKQ70UwNlJEI25uOb3/ed8gNytVM5RMBUjSlc7djkgCm0xThNJilAJkoEQ+aFYJ5Yc5FuuK5oXaSAK4BGchAWMC0I7RmilCwIowo470g7ilXG+ST2m24e8rMwemiublTOC4W+qmbDHlvc61w1o2uPyG8kKPY05ADMrZdCtHhSwjWH2r7GQ8ODOgv7V5/qOtWlw2vc+v1L4MX1JV4RLYZQMhjbFGLNaLD5k8STc3UpBDf6+skSmhueW/wDopSJgG6y+a0ekeR/UnzZ6GXLtW1HUsAab7/gE9aUg1KsK9/FjjBVFHDKTlyxwEKI0o6shDkCFdZExC6WYAytp3wPyuLtdbNrx6Lh0PuJG9eZ8RonxSuikbZ7HFrhwI+IO0HeCF6xWUdr+jbBLDXEERucyKo1bAgX8D8xa9rtvyamxPbL7P8wS5RjwHJH1VrsHZ5RN294/q+w/0gKYoNG6SH9XTsB4ka5/icSVd5UJtMfwrAaioI7qFxH3iNVg/McleME7OGNs6pfrn/LZcN83bT5WV9shU5ZGxkhGnp2RtDGNDWjY1osB5BGRiEWyQItCVI0xUZGVJUqxiRJyTKc5p3uTKYoAES5ciOcgWCeXCUOsushXaTBshvkgQgLABCMShbsRBvPDNMACR1hzKVpWAC53JuzxG6k6CidNLHCz0nkAcuJPIC5PRI5KKcn0hlzwW3s3wEzS/pLx9nGfCD6z/wCjdvW3Nauxu5McHw9sETIWeiwAdeJPU3Pmpyki3lfG5sktdqN3hdf6PVglhhXkUhisAlwgdkir28WNRVI5JSsVDko13kmxejtcr0THjClQU0a5KtciAcLkm1yPdGzAphjuFtqaeWnf6MjS2/A7Wu6ggHyT9AUTFC0dlc6nYJMpI7xSDg+ImN3t1b87qTCUxCj7uokcPRlDX/naAx/uER6kpNO+QAoFy5AwBRQEa6K1YIdikqYKOaFJUgWAx4TkmE5T55yUXVvWAN3SZrk2Lui5YJ5tXIEK7CYYI7QiBKsRQpz9lknObCyVamzjckrSMhenG9aN2VYXnJUuH4GfFx+A8ys8jaSABtO4LeNHMOEFPHDva0a3Nxzd7yV5HrOfZgWOPc/y8nXpYXPc/BMU7FK07bN65pjSRXPxUkvP9PwbI2XzzthXohKNIkSV6qOUG6OCkro10yALsKWDk0Y5La6zMOGlHDk2aUe/NAw5BRkjGlgsYaYnDrMJ3jMfP3KCsrOeCrUzdUkcCQniwBSgugQImOKBcSgBRMLxKRpVHxKTpljCsxyULWSKWq3WCr9S/NAwUOXIoauWMf/Z">
</body>
</html>
```
