### 1) 반응하는 단위들
아래 몇가지 단위들은 뷰포트 크기를 기반으로 값을 계산하여 크기를 결정하는 가변 단위입니다.


```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>다른 상대 단위들</title>
</head>
<style>
    p{
        margin: 0;
        font-size: 1vw;  /* 화면크기의 너비에서 100분의 1 너비가 달라지면 글자크기도 달라짐*/
        font-size: 50vh;  /* 화면의 높이에 따라 백분율로 글자크기가 달라짐 */
        font-size: 50vmin;  /* 너비가 높이보다 작으면 너비를 기준으로 글자크기가 달라짐(높이가 더 작으면 높이기준) */
        font-size: 50vmax;  /* 너비가 높이보다 크면 너비를 기준으로 글자크기가 달라짐(높이가 더 크면 높이기준) */
    }
</style>
<body>
    <p>왕왕</p>
</body>
</html>
```
