### 문서 기분구조

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
</head>
<body>
    
</body>
</html>
```

- viewport 관련 설정 meta 태그 <br>
    meta name="viewport" 
    content="width=device-width, initial-scale=1.0"

- "width=device-width" : 기기의 너비 기준으로 초기화하겠다.
- , initial-scale=1.0: 기존 스케일을 잡겠다.

- 해당 태그를 넣어야 모바일 버전과 PC버전에서 모두 태그의 크기 그대로 적용됨
