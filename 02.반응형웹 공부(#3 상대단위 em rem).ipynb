### 1) 절대길이 단위 px
- 입문 레벨에서 가장 많이 사용하는 단위인 px(픽셀)은 절대 길이 단위
- px은 어떤 상황에서도 동일한 값을 유지하므로, 가변성이 없다

### 2) em과 rem
- em과 rem은 박스에서 텍스트 크기를 조정할 때 사용하는 상대단위
- em : 부모 요소의 글꼴 크기
    -   em으로 내,외부 여백 크기를 정할 때는 자신의 글자 크기를 기준
- rem : 루트 요소의 글꼴 크기
    - 루트요소는 html 요소  => 루트 요소의 글자 크기 기본값은 16px (2rem  = 32px)
