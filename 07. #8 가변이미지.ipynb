# 8. 가변이미지

## 1) 이미지의 상대단위
- 이미지 요소에도 상대단위 em, vw, % 등을 얼마든지 사용할 수 있습니다.
- 유의사항
    - 브라우저가 대개 이미지의 비율을 유지한다
    - 이미지가 무작정 커지는 것을 경계해야 한다

## 2) max-width
- CSS의 max-width 속성은 너비의 최대값을 지정하여 요소의 너비가 그 이상 커지지 않도록 합니다.
- ex)
   - img { max-width: 200px; }
   - img { max-width: 100%; }  => 부모요소보다 커질 수 없음

## 3) picture
- HTML의 picture 요소를 이용하면 뷰포트의 너비 등 브라우저의 특정 조건에 따라 이미지를 반응형으로 불러올 수 있습니다.
